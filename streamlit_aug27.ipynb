```python
import getpass
import os

os.environ["GROQ_API_KEY"] = key

from langchain_groq import ChatGroq

llm = ChatGroq(model="llama3-8b-8192")
```

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

system_prompt = (
    "Consider yourself as a psychiatrist for answering the emotional state of a particular user mentioned in the prompt. "
    "Use the username field as users"
    "Also give guidance to make the emotional state better"
    "Use the following pieces of retrieved context to answer the question."
    "If you don't know the answer, say that you "
    "don't know. Use three sentences maximum and keep the "
    "answer concise."
    "\n\n"
    "{context}"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)

#retriever = vectorstore.as_retriever()
retriever = vectorstore.as_retriever()

question_answer_chain = create_stuff_documents_chain(llm, prompt)
rag_chain = create_retrieval_chain(retriever, question_answer_chain)


# List all usernames to verify data
response1 = rag_chain.invoke({"input": "List all usernames from the dataset, using the 'username' field. Ensure no users are missed."})
print("List of usernames:", response1["answer"])

# Interactive loop to continuously query the chatbot
while True:
    user = input("Enter the name of the user (or 'exit' to quit): ").strip().lower()
    if user == 'exit':
        break
    print(f"Searching for the emotional state of user: {user}")

    response = rag_chain.invoke({"input": f"What is the emotional state of the user {user}?"})
    print("Response:", response["answer"])

    if response["answer"].strip().lower() == "i don't know.":
        print("The user might not exist in the dataset or there is an issue with retrieval.")
    else:
        print("Successfully retrieved the emotional state.")
```

```python
%%writefile app.py

from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
import streamlit as st

system_prompt = (
    "Consider yourself as a psychiatrist for answering the emotional state of a particular user mentioned in the prompt. "
    "Use the username field as users"
    "Also give guidance to make the emotional state better"
    "Use the following pieces of retrieved context to answer the question."
    "If you don't know the answer, say that you "
    "don't know. Use three sentences maximum and keep the "
    "answer concise."
    "\n\n"
    "{context}"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)

#retriever = vectorstore.as_retriever()
retriever = vectorstore.as_retriever()

question_answer_chain = create_stuff_documents_chain(llm, prompt)
rag_chain = create_retrieval_chain(retriever, question_answer_chain)




# Streamlit app interface
st.title("Emotional State Chatbot")

# Input field for the username
user_input = st.text_input("Enter the name of the user:")

if st.button("Get Emotional State"):
    if user_input:
        user_input = user_input.strip().lower()
        st.write(f"Searching for the emotional state of user: {user_input}")

        # This is where the RAG chain is invoked with the user input
        response = rag_chain.invoke({"input": f"What is the emotional state of the user {user_input}?"})
        st.write("Response:", response["answer"])

        if response["answer"].strip().lower() == "i don't know.":
            st.write("The user might not exist in the dataset or there is an issue with retrieval.")
        else:
            st.write("Successfully retrieved the emotional state.")
    else:
        st.write("Please enter a username.")
```

```python
from pyngrok import ngrok

start = time.time()
# Start the Streamlit app in the background
!streamlit run app.py &
end = time.time()

print("time of execution for streamlit app:"+(end-start))
# Expose the app via ngrok
public_url = ngrok.connect(port='8501')
print(f"Streamlit app is live at {public_url}")
```
